```python
full_data['DATE'] = pd.to_datetime(full_data['DATE'])
```

get data for 2002 to 2017

```python
data02to17 = full_data.loc[(full_data.DATE.dt.year >= 2002) & (full_data.DATE.dt.year <= 2017)]
```
